```python
advertisingdf = pd.read_csv('advertising.csv', index_col=0)
```

```python
advertisingdf.head(3)
```

```python
scatter_matrix(advertisingdf, alpha=0.8, figsize=(12,12))
None
```

```python
y = advertisingdf.sales
X = advertisingdf[['TV','radio']]
#X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
model.params
model.summary()
```

```python
from itertools import combinations
results=[]
for i in range(1,4):
    combos = list(combinations(['TV', 'radio', 'newspaper'],i))
    #combos = [[*c] for c in combos]
    for com in combos:
        y = advertisingdf.sales
        X = pd.DataFrame(advertisingdf, columns=com)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
        model = linear_model.LinearRegression(fit_intercept=True).fit(X_train, y_train)
        score = model.score(X_test, y_test)
        results.append((score, com))
```

```python
answer = pd.DataFrame(results, columns=('score', 'com'))
```
