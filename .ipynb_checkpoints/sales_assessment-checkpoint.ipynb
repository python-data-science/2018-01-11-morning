```python
from itertools import combinations

total = 0
for i in range(1,11):
    total += len(list(combinations(['TV', 'radio', 'newspaper'],i)))

total
```

```python
d = {}
for i in range(1,8):
    combos = list(combinations(['TV', 'radio', 'newspaper'],i))
    #combos = [[*c] for c in combos]
    for com in combos:
        y = df.sales
        X = pd.DataFrame(df, columns=com)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True, random_state=22)
        model = linear_model.LinearRegression(fit_intercept=True).fit(X_train, y_train)
        score = model.score(X_test, y_test)
        d[score] = com
```

```python
rows = []
for i in range(1,8):
    combos = list(combinations(['TV', 'radio', 'newspaper'],i))
    for j,com in enumerate(combos):
        y = df.sales
        X = pd.DataFrame(df, columns=com)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True, random_state=22)
        model = linear_model.LinearRegression(fit_intercept=True).fit(X_train, y_train)
        score = model.score(X_test, y_test)
        s = ', '.join(com)
        rows.append({'Score':score, 'Combos':s, 'Coef':model.coef_,'Int':model.intercept_})
        # print('score:', score, 'columns:', s)
df1 = pd.DataFrame(rows)
df1.head()
```

```python
b = 3.014062 # y intercept
tv_coef = 0.0438400599475
radio_coef = 0.198307346577
news_coef = 0.0

tv_input = user_str = input('Please input a number for TV sales: ')
radio_input = user_str = input('Please input a number for radio sales: ')
newspaper_input = user_str = input('Please input a number for newspaper sales: ')

y_hat = b + (tv_coef*int(tv_input)) + (radio_coef*int(radio_input)) + (news_coef*int(newspaper_input))

print(y_hat)
```
